```python
user_key=environment_variables['eshta']
pd.set_option('display.precision', 12)
```

```python
data=pd.read_csv(user_key['extremes_raw_unclean']+'weatherstats_toronto_normals_daily-1990-2020.csv')
data2= pd.read_csv(user_key['extremes_raw_unclean']+'weatherstats-toronto-daily-1990-2020.csv')
data3= pd.read_csv(user_key['extremes_raw_unclean']+'weatherstats_toronto_normals_monthly-1990-2020.csv')
```

```python
#clean up the data here

#1. Check for nulls or zero values
#2. Check if all days are there

nullCounter = 0
zeroCounter = 0
blankCounter = 0
counter = 0
row_iterator = data.iterrows()


listofDates= []
listofColumns= []
listofMonths= []
listofYears= []


for i, row in row_iterator:
        if counter != len(data):
            for column in range(0,53):
                #print((data.at[i,data.columns[column]]))                                                                                    
                if (pd.isnull(data.at[i,data.columns[column]])):
                    #create a list of dates
                    #print('test')
                    #only append if column has _v in it
                    if ( '_v'in data.columns[column] ):
                        
                        listofDates.append(data.at[i,'date'])
                        listofMonths.append(pd.to_datetime(data.at[i,'date']).month)
                        listofYears.append(pd.to_datetime(data.at[i,'date']).year)
                        listofColumns.append(data.columns[column])
                        nullCounter += 1
                    
                    
                    
                
                    
        counter += 1   
        
        
#print(listofColumns)                        
Daily_Report_normals = pd.DataFrame()


Daily_Report_normals.insert(0, 'Date', listofDates)
Daily_Report_normals.insert(1, 'Month', listofMonths)
Daily_Report_normals.insert(2, 'Year', listofYears)
Daily_Report_normals.insert(3, 'Column', listofColumns)



Daily_Report_normals = (Daily_Report_normals.groupby(['Column', 'Year','Month']).size().reset_index(name='counts'))
```
