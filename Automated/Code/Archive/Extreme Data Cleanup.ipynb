```python
data=pd.read_csv(user_key['extremes_raw']+'extremes_raw_unclean/original files/weatherstats_toronto_normals_daily-1990-2020.csv')
data2= pd.read_csv(user_key['extremes_raw']+'extremes_raw_unclean/original files/weatherstats-toronto-daily-1990-2020.csv')
```

```python
print(data)
print(data2)
```

```python
data_final = pd.DataFrame()

data_final['date']= pd.to_datetime(data['date'])

data_final['max_temperature_normal']= data['max_temperature_v']
data_final['max_temperature_observed']= data2['max_temperature']

data_final['max_relative_humidity_normal']= data['max_relative_humidity_v']
data_final['max_relative_humidity_observed']= data2['max_relative_humidity']

data_final['min_relative_humidity_normal']= data['min_relative_humidity_v']
data_final['min_relative_humidity_observed']= data2['min_relative_humidity']

data_final['max_wind_speed_normal']= data['max_wind_speed_v']
data_final['max_wind_speed_observed']= data2['max_wind_speed']

data_final['min_wind_speed_normal']= data['min_wind_speed_v']
data_final['min_wind_speed_observed']= data2['min_wind_speed']

data_final['precipitation_normal']= data['precipitation_v']
data_final['precipitation_observed']= data2['precipitation']

data_final['rain_normal']= data['rain_v']
data_final['rain_observed']= data2['rain']

data_final['snow_normal']= data['snow_v']
data_final['snow_observed']= data2['snow']

data_final['snow_on_ground_normal']= data['snow_on_ground_v']
data_final['snow_on_ground_observed']= data2['snow_on_ground']



print(data_final)
```

```python
#clean up the data here

#1. Check for nulls or zero values
#2. Check if all days are there

nullCounter = 0
zeroCounter = 0
blankCounter = 0
counter = 0
row_iterator = data_final.iterrows()
for i, row in row_iterator:
        if counter != len(data_final):
            for column in range(0,18):
                #print(data_final.at[i,data_final.columns[column]])
                if (data_final.at[i,data_final.columns[column]] == ''):
                    blankCounter += 1
                if (data_final.at[i,data_final.columns[column]] == 0):
                    zeroCounter += 1
                    
                
        counter += 1        
Daily_Report = pd.DataFrame()
NullValues = []
#Daily_Report['date']= pd.to_datetime(data['date'])

for column in range(0,18):
    NullValues.append(data_final.columns[column] + " : " + str(data_final[data_final.columns[column]].isna().sum()))
    
    
    

        
#nullCounter = data_final.isnull().sum().sum()

print(nullCounter)
print(blankCounter)
print(zeroCounter)
Daily_Report.insert(0, 'Null per coulmn', NullValues)
Daily_Report.insert(1, 'Total Blanks', blankCounter)
Daily_Report.insert(2, 'Total Zero Value', zeroCounter)
```

```python
data_final.to_csv(user_key['extremes_raw']+'extremes_raw_clean/daily_normals.csv', index=None)
Daily_Report.to_csv(user_key['extremes_raw']+'extremes_raw_clean/daily_report.csv', index=None)
```
