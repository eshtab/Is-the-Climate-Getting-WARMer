```python
import pandas as pd
import datetime as dt
import numpy as np
import csv
import calendar
from pprint import pprint

import dash
import dash_core_components as dcc
import dash_html_components as html

import os,sys,inspect
from dash.dependencies import Input, Output

current_dir = os.path.dirname(os.path.abspath(inspect.getfile(inspect.currentframe())))
#parent_dir = os.paths.dirname(current_dir)
sys.path.insert(0, current_dir) 
from environmentVariables import environment_variables
```

```python
user_key=environment_variables['mahaveer']
pd.set_option('display.precision', 12)
```

```python
#we want to convert the WRDC file, which has data seperated, and consolidate into one sheet
#each sheet has a column DATE going range 1-31, and individual columns for month, we need to traverse
dates = []
df = pd.DataFrame(columns=['Date','rsds'])
counter = 0

for i in range(2,13):
    data=pd.read_excel(user_key['GSOD_Loc']+'WRDC.xlsx',sheet_name=i)
    year = 0
    if i == 2:
        year = 1990
    elif i == 3:
        year = 1991
    elif i == 4:
        year = 1992
    elif i == 5:
        year = 1993        
    elif i == 6:
        year = 1994       
    elif i == 7:
        year = 1995           
    elif i == 8:
        year = 1996   
    elif i == 9:
        year = 1997        
    elif i == 10:
        year = 1998        
    elif i == 11:
        year = 1999  
    elif i == 12:
        year = 2000        
    data.columns = data.iloc[0]
    data = data[1:]
    row_ittr2 = data.iterrows()

    for i, row in row_ittr2:
        
        if(isinstance(data.at[i,'DATE'], int)):
            
        
            #we have to do all columns same time
            column = 'JAN'

            if (data.at[i,'JAN'] ):
                month = 1
                date = dt.datetime(year, month, data.at[i,'DATE'])
                dates.append(date)
                df.loc[counter] = [date,data.at[i,'JAN']]
                counter += 1
                
            
            if (calendar.isleap(year) and data.at[i,'FEB'] ):
                if (data.at[i,'DATE'] < 30):    
                    month = 2
                    date = dt.datetime(year, month, data.at[i,'DATE'])
                    dates.append(date)
                    df.loc[counter] = [date,data.at[i,'FEB']]
                    counter += 1
            else:
                if (data.at[i,'DATE'] < 29):
                    month = 2
                    date = dt.datetime(year, month, data.at[i,'DATE'])
                    dates.append(date)
                    df.loc[counter] = [date,data.at[i,'FEB']]
                    counter += 1
            
            if (data.at[i,'MAR'] ):
                month = 3
                date = dt.datetime(year, month, data.at[i,'DATE'])
                dates.append(date)
                df.loc[counter] = [date,data.at[i,'MAR']]
                counter += 1
            if (data.at[i,'APR'] and data.at[i,'DATE'] < 31):
                month = 4
                date = dt.datetime(year, month, data.at[i,'DATE'])
                dates.append(date)
                df.loc[counter] = [date,data.at[i,'APR']]
                counter += 1
            if (data.at[i,'MAY'] ):
                month = 5
                date = dt.datetime(year, month, data.at[i,'DATE'])
                dates.append(date)
                df.loc[counter] = [date,data.at[i,'MAY']]
                counter += 1
            if (data.at[i,'JUN'] and data.at[i,'DATE'] < 31):
                month = 6
                date = dt.datetime(year, month, data.at[i,'DATE'])
                dates.append(date)
                df.loc[counter] = [date,data.at[i,'JUN']]
                counter += 1
            if (data.at[i,'JUL'] ):
                month = 7
                date = dt.datetime(year, month, data.at[i,'DATE'])
                dates.append(date)
                df.loc[counter] = [date,data.at[i,'JUL']]
                counter += 1
            if (data.at[i,'AUG'] ):
                month = 8
                date = dt.datetime(year, month, data.at[i,'DATE'])
                dates.append(date)
                df.loc[counter] = [date,data.at[i,'AUG']]
                counter += 1
            if (data.at[i,'SEP'] and data.at[i,'DATE'] < 31 ):
                month = 9
                date = dt.datetime(year, month, data.at[i,'DATE'])
                dates.append(date)
                df.loc[counter] = [date,data.at[i,'SEP']]
                counter += 1
            if (data.at[i,'OCT'] ):
                month = 10
                date = dt.datetime(year, month, data.at[i,'DATE'])
                dates.append(date)
                df.loc[counter] = [date,data.at[i,'OCT']]
                counter += 1
            if (data.at[i,'NOV'] and data.at[i,'DATE'] < 31):
                month = 11
                date = dt.datetime(year, month, data.at[i,'DATE'])
                dates.append(date)
                df.loc[counter] = [date,data.at[i,'NOV']]
                counter += 1
            if (data.at[i,'DEC'] ):
                month = 12
                date = dt.datetime(year, month, data.at[i,'DATE'])
                dates.append(date)
                df.loc[counter] = [date,data.at[i,'DEC']]
                counter += 1
            
            
#dates = sorted(dates)
#print('\n'.join(str(e) for e in dates))
print(df)
```

```python
#sort the dataframe by date
df = df.sort_values(by='Date')
print(df)
```
